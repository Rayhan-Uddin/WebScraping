```python
data = []
url = f"https://www2.hm.com/en_us/men/products/pants/_jcr_content/main/productlisting.display.json?sort=stock&image-size=small&image=model&offset=36&page-size=358"
req = requests.get(url, headers=headers)
js = req.json()

for product in range(1, 358):
    d = {
        'Product_name': js["products"][product]["title"],  # attrbt 1
        'Product_catagory': js["products"][product]["category"],  # atrbt 2
        'price': js["products"][product]["price"].split(" ")[1],  # atrtbt 4
        'redprice': js["products"][product]["redPrice"],  # atrtbt 5
        'yellowPrice': js["products"][0]["yellowPrice"],  # atrtbt 6
        'blue_price': js["products"][product]["bluePrice"],  # atrtbt 7
        'selling_attribute': js["products"][product]["sellingAttribute"],
        'Brand_Name': js["products"][product]["brandName"],  # atrtbt 8
        # atribute 9
        'color_available': js["products"][product]["swatchesTotal"],
        'product_link': "https://www2.hm.com/" + js["products"][0]["link"]
    }
    print("__________")
    print(d)
    print("__________")
    data.append(d)

```

getting all the data

```python
import pandas as pd 
df= pd.DataFrame(data)
df.head()
```
